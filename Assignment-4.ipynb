# importing packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

#reading as loading data

```python
data=pd.read_csv('/content/50_Startups.csv')
data.sample(20)
```

checking for missing values

```python
data.isnull().sum()
```

Deleting profit column from dataset consider as y
and remaining columns are consider as x

```python

x=data.drop("Profit",axis=1).values
y=data["Profit"].values
```

```python
x.shape
```

```python
y.shape
```

```python
y.reshape(-1,1).shape  # reshaping y data for testing and traing module
```

```python
regre=LinearRegression()
regre.fit(x,y)
```

```python

y_pred=regre.predict(x) #regression predected values stored in y_pred
```

```python

print("MAE:",mean_absolute_error(y,y_pred))
print("MSE:",mean_squared_error(y,y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y,y_pred)))
print("R2 Score:",r2_score(y,y_pred))
```

 **L1 regularization**

*Splecting data into train 70% and test 30%*

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python

l=Lasso(alpha=0.1)
l.fit(x_train,y_train)
```

Predected X values is stored in y_pred_losso

```python

y_pred_lasso=l.predict(x_train)
```

##Model with Training data

```python

print("MAE:",mean_absolute_error(y_train,y_pred_lasso))
print("MSE:",mean_squared_error(y_train,y_pred_lasso))
print("RMSE:",np.sqrt(mean_squared_error(y_train,y_pred_lasso)))
print("R2 Score:",r2_score(y_train,y_pred_lasso))
```

##Model with Testing data

```python

y_pred2_lasso=l.predict(x_test)
print("MAE:",mean_absolute_error(y_test,y_pred2_lasso))
print("MSE:",mean_squared_error(y_test,y_pred2_lasso))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred2_lasso)))
print("R2 Score:",r2_score(y_test,y_pred2_lasso))
```

**L2 regularization**

```python

r=Ridge(alpha=0.1)
r.fit(x_train,y_train)
```

```python

y_pred_ridge=r.predict(x_train)
```

Model with training data

```python

print("MAE:",mean_absolute_error(y_train,y_pred_ridge))
print("MSE:",mean_squared_error(y_train,y_pred_ridge))
print("RMSE:",np.sqrt(mean_squared_error(y_train,y_pred_ridge)))
print("R2 Score:",r2_score(y_train,y_pred_ridge))
```

```python

y_pred2_ridge=r.predict(x_test)
```

Model with testing data

```python

print("MAE:",mean_absolute_error(y_test,y_pred2_ridge))
print("MSE:",mean_squared_error(y_test,y_pred2_ridge))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred2_ridge)))
print("R2 Score:",r2_score(y_test,y_pred2_ridge))
```
