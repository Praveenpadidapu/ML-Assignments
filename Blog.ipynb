```python
import numpy as np
import pandas as pd
data=pd.read_csv('/content/world_health_data.csv')
data
```

```python
data.isnull().sum()
```

```python
mean=data.mean().mean()
mean
```

```python
data.columns
```

```python
columns = ['health_exp', 'life_expect',
       'maternal_mortality', 'infant_mortality', 'neonatal_mortality',
       'under_5_mortality', 'prev_hiv', 'inci_tuberc',
       'prev_undernourishment']
mean=data[columns].mean()
mean
```

```python
data['life_expect'].fillna(data['life_expect'].mean(),inplace=True)
data['health_exp'].fillna(data['health_exp'].mean(),inplace=True)
data['maternal_mortality'].fillna(data['maternal_mortality'].mean(),inplace=True)
data['infant_mortality'].fillna(data['infant_mortality'].mean(),inplace=True)
data['neonatal_mortality'].fillna(data['neonatal_mortality'].mean(),inplace=True)
data['under_5_mortality'].fillna(data['under_5_mortality'].mean(),inplace=True)
data['prev_hiv'].fillna(data['prev_hiv'].mean(),inplace=True)
data['inci_tuberc'].fillna(data['inci_tuberc'].mean(),inplace=True)
data['prev_undernourishment'].fillna(data['prev_undernourishment'].mean(),inplace=True)
```
