```python
plt.figure(figsize=(8, 6))
plt.hist(data['Units_Sold'], bins=20, color='skyblue', edgecolor='black')  # Create histogram
plt.title('Distribution of Units sold')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

##Number of units sold for variours Product Category

```python
plt.figure(figsize=(8, 6))
data['Product_Category'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Units_sold')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

**Number of Units sold on customer segment**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Customer_Segment', y='Units_Sold', data=data, palette='Set3')
plt.title('Units Sold Distribution by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Units Sold')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

# **Observation** : as per Box plot it clearly seems to be understand that Occasional segment as more out layer  and one out layer in Premium Customer Segment

# Distribution of Units Sold by Customer Segment

```python
import pandas as pd
import matplotlib.pyplot as plt
segment_units = data.groupby('Customer_Segment')['Units_Sold'].sum()
plt.figure(figsize=(8, 6))
plt.pie(segment_units, labels=segment_units.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral','lightgreen'])
plt.title('Distribution of Units Sold by Customer Segment')
plt.show()
```

# **Observation**:above pie chart shows data about percentage of units sold on various Customer segment .All segments has almost same percentage

```python
import pandas as pd
import matplotlib.pyplot as plt
price_bins = [0, 200, 400, 600, 800, 1000]
data['Price Range'] = pd.cut(data['Price'], bins=price_bins)
units_sold_by_price = data.groupby('Price Range')['Units_Sold'].sum()
plt.figure(figsize=(8, 6))
units_sold_by_price.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Total Units Sold by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```
